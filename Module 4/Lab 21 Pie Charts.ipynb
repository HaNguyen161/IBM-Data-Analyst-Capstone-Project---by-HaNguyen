**Download and Load the Data**


To start, download and load the dataset into a `pandas` DataFrame.



```python
# Step 1: Download the dataset
!wget -O survey-data.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv

# Step 2: Import necessary libraries and load the dataset
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("survey-data.csv")

# Display the first few rows to understand the structure of the data
df.head()
```

### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
# Task 1.1: Create a Pie Chart of the Top 5 Databases Respondents Want to Work With

# Count the top 5 databases
top_databases = df['DatabaseWantToWorkWith'].value_counts().head(5)

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_databases.values,
    labels=top_databases.index,
    autopct='%1.1f%%',
    startangle=90
)

# Add title
plt.title('Top 5 Databases Respondents Want to Work With')

# Display chart
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
# Task 1.2: Create a Pie Chart of the Top 5 Developer Roles

# Count the top 5 developer types
top_devtypes = df['DevType'].value_counts().head(5)

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_devtypes.values,
    labels=top_devtypes.index,
    autopct='%1.1f%%',
    startangle=90
)

# Add title
plt.title('Top 5 Developer Roles')

# Display chart
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use
