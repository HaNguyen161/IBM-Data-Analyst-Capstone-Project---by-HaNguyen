## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Write your code here
# Find duplicated rows
duplicates = df.duplicated(keep=False)

# Count duplicate entries (excluding first occurrence)
duplicate_count = df.duplicated().sum()

print(f"Number of duplicate rows: {duplicate_count}")

print("\nFirst few duplicate rows:")
print(df[duplicates].head())
```

### Task 2: Analysis of Duplicate Response Patterns
1. Identify duplicate response patterns based on selected columns such as MainBranch, Employment, and RemoteWork.
2. Clarify that these represent multiple respondents with identical answers rather than duplicate records. Analyse which other columns frequently share identical values within these response-pattern groups.
   


```python
# Columns defining the response pattern
subset_cols = ['MainBranch', 'Employment', 'RemoteWork']

# Find all respondents sharing these patterns
pattern_groups = df[df.duplicated(subset=subset_cols, keep=False)]

print("These are shared response patterns, not duplicate records.\n")

# Analyze groups
for pattern, group in pattern_groups.groupby(subset_cols):

    # Skip very small groups
    if len(group) <= 5:
        continue

    print(f"\nPattern: {pattern}")
    print(f"Respondents: {len(group)}")

    # Find columns where everyone answered identically
    shared_cols = {
        col: group[col].iloc[0]
        for col in df.columns
        if col not in subset_cols and group[col].nunique() == 1
    }

    if shared_cols:
        print("Shared identical answers:")
        for col, val in shared_cols.items():
            print(f" - {col}: {val}")
    else:
        print("No additional identical columns found.")
```

### Task 3: Visualize Shared Response Patterns
1. Create visualizations to show the distribution of shared response patterns across different categories.
2. Use bar charts or pie charts to represent the distribution of respondents who share identical values for MainBranch, Employment, and RemoteWork, grouped by Country and Employment.


```python
!pip install seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Columns used to identify shared response patterns
pattern_cols = ['MainBranch', 'Employment', 'RemoteWork']

# Find rows with duplicated response patterns
shared = df[df.duplicated(subset=pattern_cols, keep=False)]

# Count respondents by Country and Employment
dist = (
    shared.groupby(['Country', 'Employment'])
    .size()
    .reset_index(name='Count')
    .sort_values('Count', ascending=False)
    .head(15)
)

# ----- Bar Chart -----
plt.figure(figsize=(12, 8))

sns.barplot(
    data=dist,
    x='Count',
    y='Country',
    hue='Employment'
)

plt.title('Top Shared Response Patterns by Country and Employment')
plt.tight_layout()
plt.show()

# ----- Pie Chart -----
# Employment distribution among shared patterns
emp_dist = shared['Employment'].value_counts()

# Keep top 4 categories, group remaining into "Other"
top4 = emp_dist.head(4)

if len(emp_dist) > 4:
    top4['Other'] = emp_dist.iloc[4:].sum()

plt.figure(figsize=(8, 8))

plt.pie(
    top4,
    labels=top4.index,
    autopct='%1.1f%%'
)

plt.title('Employment Distribution of Shared Patterns')
plt.show()
```
