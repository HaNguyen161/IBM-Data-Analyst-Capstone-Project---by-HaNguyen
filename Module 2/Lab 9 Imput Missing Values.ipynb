### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**
<p>
The functions below will download the dataset into your browser:
</p>


```python
file_path ="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/UDKAZw-kz18Yj8P6icf_qw/survey-data-duplicates.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
## Write your code here
# 1. Identify if rows are duplicates (returns a boolean Series True/False)
duplicates_boolean = df.duplicated()

# 2. View the actual duplicate rows in the dataset
duplicate_rows = df[df.duplicated()]
print("Duplicate Rows:")
print(duplicate_rows)

# 3. Count the total number of duplicate rows
num_duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {num_duplicates}")
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
## Write your code here

# Assuming 'df' is your existing DataFrame
# This creates a new DataFrame with duplicate rows removed:
df_cleaned = df.drop_duplicates()

# If you want to remove duplicates from the original DataFrame directly, use inplace=True:
df.drop_duplicates(inplace=True)
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
## Write your code here

# Assuming you have already loaded your dataset into a DataFrame named 'df'
# This will calculate the total number of missing values in each column
missing_values = df.isnull().sum()

# Alternatively, you can use isna(), which does the exact same thing:
# missing_values = df.isna().sum()

print("Missing values per column:")
print(missing_values)
```

##### Task 4: Find out how many rows are missing in the column RemoteWork.


```python
## Write your code here

# Assuming you have loaded your dataset into a DataFrame called 'df'
missing_remote_work = df['RemoteWork'].isnull().sum()

# Alternatively: df['RemoteWork'].isna().sum()

print(f"Missing values in RemoteWork column: {missing_remote_work}")
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
## Write your code here

# Assuming you have loaded your dataset into a DataFrame called 'df'
# This will return the frequency of each unique value in the 'RemoteWork' column
remote_work_counts = df['RemoteWork'].value_counts()

# If you also want to see the count of missing (NaN) values included in the results:
# remote_work_counts_with_nan = df['RemoteWork'].value_counts(dropna=False)

print("Value counts for RemoteWork:")
print(remote_work_counts)
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
## Write your code here
most_frequent_value = df['RemoteWork'].mode()

print(f"The most frequent value in RemoteWork is: {most_frequent_value}")
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
## Write your code here
# 1. Identify the most frequent (majority) value in the RemoteWork column
# The mode() function returns a Series, so we use  to extract the actual value
majority_value = df['RemoteWork'].mode()

# 2. Replace (impute) the missing/empty values with this majority value
# Using inplace=True applies the change directly to your existing DataFrame
df['RemoteWork'].fillna(majority_value, inplace=True)

print(f"Empty rows in RemoteWork have been replaced with: {majority_value}")
```

##### Task 8: Check for any compensation-related columns and describe their distribution.



```python
## Write your code here
# 1. Identify compensation-related columns using keywords
keywords = ['comp', 'salary', 'pay', 'wage', 'income']
comp_columns = [col for col in df.columns if any(keyword in col.lower() for keyword in keywords)]

print(f"Compensation-related columns found: {comp_columns}\n")

# 2. Describe the distribution of these columns
if comp_columns:
    for col in comp_columns:
        print(f"--- Distribution for {col} ---")
        
        # .describe() provides count, mean, std, min, max, and quartiles for numeric data
        distribution_stats = df[col].describe()
        print(distribution_stats)
        print("\n")
else:
    print("No compensation-related columns were found in the dataset.")
```
