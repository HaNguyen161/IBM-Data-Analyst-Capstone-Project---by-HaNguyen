#### Step 2: Load and Preview the Dataset
Load the dataset from the provided URL. Use df.head() to display the first few rows to get an overview of the structure.

#### Step 3: Handling Missing Data


Identify and manage missing values in critical columns such as `Employment`, `JobSat`, and `RemoteWork`. Implement a strategy to fill or drop these values, depending on the significance of the missing data.


```python
# -------------------------------
# 1. Check missing values
# -------------------------------
critical_columns = ["Employment", "JobSat", "RemoteWork"]

print("Missing values before handling:\n")
print(df[critical_columns].isnull().sum())

# -------------------------------
# 2. Handle missing values
# -------------------------------

# Employment:
# Important categorical column → fill with mode
employment_mode = df["Employment"].mode()[0]
df["Employment"] = df["Employment"].fillna(employment_mode)

# RemoteWork:
# Also categorical → fill with mode
remotework_mode = df["RemoteWork"].mode()[0]
df["RemoteWork"] = df["RemoteWork"].fillna(remotework_mode)

# JobSat:
# Satisfaction score is more subjective.
# If missing percentage is small, dropping rows is safer.
df = df.dropna(subset=["JobSat"])

# -------------------------------
# 3. Verify results
# -------------------------------
print("\nMissing values after handling:\n")
print(df[critical_columns].isnull().sum())

# -------------------------------
# 4. Optional: Dataset shape
# -------------------------------
print("\nDataset shape after cleaning:")
print(df.shape)
```

#### Step 4: Analysis of Experience and Job Satisfaction


Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
# -----------------------------------
# 1. Keep necessary columns
# -----------------------------------
df = df[["YearsCodePro", "JobSat"]]

# -----------------------------------
# 2. Clean YearsCodePro
# -----------------------------------
# Convert special text values if they exist
df["YearsCodePro"] = df["YearsCodePro"].replace({
    "Less than 1 year": 0.5,
    "More than 50 years": 51
})

# Convert to numeric
df["YearsCodePro"] = pd.to_numeric(df["YearsCodePro"], errors="coerce")

# Convert JobSat to numeric if needed
df["JobSat"] = pd.to_numeric(df["JobSat"], errors="coerce")

# Remove missing values
df = df.dropna(subset=["YearsCodePro", "JobSat"])

# -----------------------------------
# 3. Create experience ranges
# -----------------------------------
bins = [0, 5, 10, 20, 100]
labels = ["0-5 years", "5-10 years", "10-20 years", ">20 years"]

df["ExperienceRange"] = pd.cut(
    df["YearsCodePro"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# -----------------------------------
# 4. Calculate median JobSat
# -----------------------------------
median_jobsat = df.groupby("ExperienceRange")["JobSat"].median()

print("Median Job Satisfaction by Experience Range:\n")
print(median_jobsat)

# -----------------------------------
# 5. Visualization
# -----------------------------------
plt.figure(figsize=(8,5))

median_jobsat.plot(kind="bar")

plt.title("Median Job Satisfaction by Experience Level")
plt.xlabel("Experience Range")
plt.ylabel("Median Job Satisfaction")
plt.xticks(rotation=0)

plt.show()
```

#### Step 7: Analyzing Programming Language Trends by Region


Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(data_url)

# -----------------------------------
# 1. Keep relevant columns
# -----------------------------------
lang_df = df[["Country", "LanguageHaveWorkedWith"]].dropna()

# -----------------------------------
# 2. Select regions/countries to analyze
# -----------------------------------
selected_countries = [
    "United States of America",
    "India",
    "Germany",
    "United Kingdom of Great Britain and Northern Ireland"
]

lang_df = lang_df[
    lang_df["Country"].isin(selected_countries)
]

# -----------------------------------
# 3. Split languages into rows
# -----------------------------------
lang_df["LanguageHaveWorkedWith"] = (
    lang_df["LanguageHaveWorkedWith"]
    .str.split(";")
)

lang_exploded = lang_df.explode("LanguageHaveWorkedWith")

# -----------------------------------
# 4. Count language usage by country
# -----------------------------------
language_counts = (
    lang_exploded
    .groupby(["Country", "LanguageHaveWorkedWith"])
    .size()
    .reset_index(name="Count")
)

# -----------------------------------
# 5. Get top 5 languages per country
# -----------------------------------
top_languages = (
    language_counts
    .sort_values(["Country", "Count"], ascending=[True, False])
    .groupby("Country")
    .head(5)
)

print(top_languages)

# -----------------------------------
# 6. Visualization
# -----------------------------------
for country in selected_countries:

    country_data = top_languages[
        top_languages["Country"] == country
    ]

    plt.figure(figsize=(8,5))

    plt.bar(
        country_data["LanguageHaveWorkedWith"],
        country_data["Count"]
    )

    plt.title(f"Top Programming Languages in {country}")
    plt.xlabel("Programming Language")
    plt.ylabel("Number of Developers")
    plt.xticks(rotation=45)

    plt.show()
```

#### Step 8: Correlation Between Experience and Satisfaction


Examine how years of experience (`YearsCodePro`) correlate with job satisfaction (`JobSatPoints_1`). Use a scatter plot to visualize this relationship.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(data_url)

# -----------------------------------
# 1. Keep relevant columns
# -----------------------------------
corr_df = df[["YearsCodePro", "JobSatPoints_1"]]

# -----------------------------------
# 2. Clean YearsCodePro
# -----------------------------------
corr_df["YearsCodePro"] = corr_df["YearsCodePro"].replace({
    "Less than 1 year": 0.5,
    "More than 50 years": 51
})

# Convert columns to numeric
corr_df["YearsCodePro"] = pd.to_numeric(
    corr_df["YearsCodePro"],
    errors="coerce"
)

corr_df["JobSatPoints_1"] = pd.to_numeric(
    corr_df["JobSatPoints_1"],
    errors="coerce"
)

# -----------------------------------
# 3. Remove missing values
# -----------------------------------
corr_df = corr_df.dropna()

# -----------------------------------
# 4. Calculate correlation
# -----------------------------------
correlation = corr_df["YearsCodePro"].corr(
    corr_df["JobSatPoints_1"]
)

print("Correlation between experience and satisfaction:")
print(correlation)

# -----------------------------------
# 5. Scatter plot
# -----------------------------------
plt.figure(figsize=(8,5))

plt.scatter(
    corr_df["YearsCodePro"],
    corr_df["JobSatPoints_1"],
    alpha=0.5
)

plt.title("Experience vs Job Satisfaction")
plt.xlabel("Years of Professional Coding Experience")
plt.ylabel("Job Satisfaction (JobSatPoints_1)")

plt.show()
```

#### Step 9: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(data_url)

# -----------------------------------
# 1. Keep relevant columns
# -----------------------------------
edu_emp_df = df[["EdLevel", "Employment"]].dropna()

# -----------------------------------
# 2. Create cross-tabulation
# -----------------------------------
cross_tab = pd.crosstab(
    edu_emp_df["EdLevel"],
    edu_emp_df["Employment"]
)

print("Cross-tabulation:\n")
print(cross_tab)

# -----------------------------------
# 3. Normalize by row (percentage)
# -----------------------------------
cross_tab_percent = pd.crosstab(
    edu_emp_df["EdLevel"],
    edu_emp_df["Employment"],
    normalize="index"
) * 100

print("\nPercentage distribution:\n")
print(cross_tab_percent.round(2))

# -----------------------------------
# 4. Visualization
# -----------------------------------
cross_tab_percent.plot(
    kind="bar",
    stacked=True,
    figsize=(12,7)
)

plt.title("Employment Type by Educational Background")
plt.xlabel("Education Level")
plt.ylabel("Percentage of Respondents")

plt.legend(
    title="Employment Type",
    bbox_to_anchor=(1.05, 1),
    loc="upper left"
)

plt.tight_layout()
plt.show()
```
